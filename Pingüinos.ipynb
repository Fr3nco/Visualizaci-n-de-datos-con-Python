# DATOS DE PINGÜINOS

350 INDIVIDUOS

caracteristicas:

Longitud del pico,Profundidad del pico,Longitud de la aleta, Masa corporal, Especie,Isla,Sexo

```python
#["Longitud_pico","Profundidad_pico","Longitud_aleta","Masa", "Especie","Isla","Sexo"]
import matplotlib.pyplot as plt
import seaborn as sns

#El conjunto de datos esta precargado en seaborn

#sns.get_dataset_names() Este es para ver los conjuntos de datos que hay para utilizar
pinguinos = sns.load_dataset("penguins") #dataset

pinguinos.columns =["Especie","Isla","Long_pico","Profundidad_pico", "Long_aleta","Masa","Sexo"]

pinguinos


```

# Primera Gráfica de Seaborn

Parametros ajustables

style: white, dark, whitegrid, darkgrid,ticks

context: paper, notebook, talk, poster

```python
sns.set_theme(style="darkgrid", context="talk")#ajuste de tema
relacional = sns.relplot(data=pinguinos, height= 7, aspect = 1.2, markers = ["s", ".", "p", "d"]) # pinto sobre el lienzo, height agranda, aspect alarga

relacional.set(title="Analisis de Pingüinos")
relacional.set_ylabels("Variables")
relacional.set_xlabels("Número de Pingüino")
plt.grid(False) #por si queres sacar las rejillas
plt.show() # la muestro

#Al ser una gráfica de dispersión solo toma los datos númericos
```
