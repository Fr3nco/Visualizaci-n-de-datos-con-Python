# Examen de certificacion para estudiantes de 5 niveles

```python
import numpy as np

años = ["2010","2011","2012","2013","2014","2015","2016","2017","2018","2019"]

nivel1 = np.random.rand(10)*100
nivel2 = np.random.rand(10)*200 + 100
nivel3 = np.random.rand(10)*300 + 200
nivel4 = np.random.rand(10)*400 + 300
nivel5 = np.random.rand(10)*500 + 400




```

# Personalizacion de graficos

```python
#--------------------------------------------------------------------------Introducir series de datos con color, marcador, etiqueta y estilo
#En el mismo parentesis se agregan los atributos para personalizar el plot
plt.plot(años,nivel1, label = "nivel 1", color = "purple", marker='o', linestyle = "-")
plt.plot(años,nivel2, label = "nivel 2", color = "red", marker='*', linestyle = "--")
plt.plot(años,nivel3, label = "nivel 3", color = "orange", marker='>', linestyle = ":")
plt.plot(años,nivel4, label = "nivel 4", color = "green", marker='<', linestyle = "-.")
plt.plot(años,nivel5, label = "nivel 5", color = "blue", marker='.', linestyle = " ")

#--------------------------------------------------------------------------activar leyendas
plt.legend() # Esto es para activar la etiqueta label
#--------------------------------------------------------------------------titulo de grafica
plt.title("Niveles de Certificación")
#--------------------------------------------------------------------------titulo de ejes
plt.xlabel("Años que se realizó el examen")
plt.ylabel("Puntaje")

#--------------------------------------------------------------------------Personalizar eje de puntaje de 0 a 1000 en incremento de 50
plt.yticks(np.arange(0,1001,50)) #Se le dicen ticks a los valores del eje y
#--------------------------------------------------------------------------activar cuadricula
plt.grid()
#--------------------------------------------------------------------------activar marcas menores
plt.minorticks_on()

plt.show()
```

# GRAFICA DE BARRAS

```python
eje_x = np.arange(0,10) # creo un eje x aleatorio de numeros del 0 al 10
eje_y = np.random.rand(10)*300
plt.bar(eje_x,eje_y)
plt.show()

eje_x = np.arange(0,10)

#asi estan solapados
plt.bar(eje_x,nivel5)
plt.bar(eje_x,nivel4)
plt.bar(eje_x,nivel3)
plt.bar(eje_x,nivel2)
plt.bar(eje_x,nivel1)
plt.show()

#para acomodarlos el plt.bar tiene que ir acompañado de unos atributos especiales
#1) se defasan el eje x
#2) luego se reajusta el ancho de la barra
plt.bar(eje_x,nivel5, width=1/5) #tomo 1/5 ya que son 5 niveles
plt.bar(eje_x+0.2,nivel4, width=1/5)
plt.bar(eje_x+0.4,nivel3, width=1/5)
plt.bar(eje_x+0.6,nivel2, width=1/5)
plt.bar(eje_x+0.8,nivel1, width=1/5)
plt.show()
```

# GRAFICA DE BARRAS HORIZONTALES(barh = bar (barra) + h(horizontal))


```python
#En este se usa height en vez del width
plt.barh(eje_x,nivel5, height=1/5)
plt.barh(eje_x+0.2,nivel4, height=1/5)
plt.barh(eje_x+0.4,nivel3, height=1/5)
plt.barh(eje_x+0.6,nivel2, height=1/5)
plt.barh(eje_x+0.8,nivel1, height=1/5)
plt.show()
```

# Grafica de dispersión (scatter)


```python
plt.scatter(eje_x,nivel5)
plt.scatter(eje_x,nivel4)
plt.scatter(eje_x,nivel3)
plt.scatter(eje_x,nivel2)
plt.scatter(eje_x,nivel1)
plt.show()
```

# Gráfica de pie (torta)

```python
plt.pie(nivel1,labels = ["Argentina","Mexico","Uruguay","Chile","Honduras","EEUU", "Mexico","Uruguay","Chile","Honduras"])
plt.show()
```

# COMBINACION DE GRÁFICAS

```python
plt.bar(eje_x,nivel2, width=1/5)
plt.scatter(eje_x,nivel1)
plt.scatter(eje_x,nivel4)

plt.plot(años,nivel5)
plt.show()
```
